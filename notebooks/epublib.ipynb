```python
import ebooklib
from ebooklib import epub
from bs4 import BeautifulSoup, Comment


books = [
    epub.read_epub("/Users/tsimur.hadeliya/code/_other/belarusian_poetry/belarussian-poetry/scripts/poems/Адамовіч Славамір/epubs/Alzyr.epub")
]
```

```python
for book in books:
    docs = book.get_items_of_type(ebooklib.ITEM_DOCUMENT)

    poem_lines = []
    for doc in docs:
        # parse each one as HTML
        soup = BeautifulSoup(doc.content, 'html.parser')

        # if the poem is wrapped in a <div class="POETRY">, narrow to that
        poetry_div = soup.find('div', class_='POETRY')
        target = poetry_div or soup.body

        # extract all the <p> tags
        for p in target.find_all('p'):
            text = p.get_text(strip=True)
            if text:
                poem_lines.append(text)

    # 3) join and print
    print('\n'.join(poem_lines))
    print("-------------=========")
    for item in book.get_items():
        if item.get_type() == ebooklib.ITEM_DOCUMENT:
            print(BeautifulSoup(item.get_content(), features="xml"))    
    print("++++++++++++++++++")
```
